```python
# *args and **kwargs Example
# *args allows you to pass a variable number of non-keyword arguments to a function. Received as tuple
# **kwargs allows you to pass a variable number of keyword arguments to a function. Received as dictionary
def create_invite(event_name, *guest_names, **details):
    print(f"Event: {event_name}\n")

    print("Guests:")
    for guest in guest_names:
        print(f"- {guest}")

    print("\nEvent Details:")
    for key, value in details.items():
        print(f"{key}: {value}")



create_invite(
    "Birthday Party",
    "Alice",
    "Bob",
    "Charlie",
    Time="6 PM",
    Location="Bangalore",
    DressCode="Casual"
)

```
