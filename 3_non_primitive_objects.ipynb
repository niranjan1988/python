```python
# List Operations in Python
a=[1,'name', 10.00]
a.append(25)
print(a)
a.append('hello')
print(a)
a.pop()
print(a)
a.pop(0)
print(a)
a.insert(0, 'first')
a.remove(10.00)
print(a)
```

```python
# Dictionary Operations in Python
d={'name':'John', 'age':30, 'city':'New York'}
print(d)
d['address']='5th Avenue'
print(d)
print(d['name'])
print(d.get('name'))
print (d.keys())
print (d.values())
```

```python
# Tuple Operations in Python
t=(1,'name', 10.00)
# Tuples are immutable, so we cannot add or remove elements
# Below operations will raise errors
t.append(25)

# append, pop, insert, remove are not valid for tuples
```

```python
# set Operations in Python  
a={1,2,3,4}
a.add(5)
a.add(5)  # Duplicate, will have no effect
print(a)
a.remove(6)  # Will raise KeyError
```

```python
# Using enumerate in Python
a=[1,2,3,4]
for item in enumerate(a):
    print(item)
    print(item[0], item[1])
```
