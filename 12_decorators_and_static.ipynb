# Concept : Decorator

```python
import time

# Step-1 : Create a decorator

def my_decorator(func):

    def wrapper():
        t1 = time.perf_counter()
        func()
        t2 = time.perf_counter()
        total_time_takes = t2-t1
        print(f"Function took {total_time_takes:.2f} seconds")
    return wrapper

# Step-2 : Create base function
@my_decorator
def loop_and_multiply():
    total = 1
    for i in range(1,100):
        total += i
        time.sleep(0.01)
    print(total)

# Step-3 : Decorate base method with decoratori.e. add @my_decorator on top of your base function
# Step-4 : Call the base function and it will execute the wrapper function
loop_and_multiply()

```

# Decorator Final concept

```python
def decorator_exep_v2(func):
    def wrapper(*args, **kwargs):
        try:
            func(*args,**kwargs)
        except Exception as e:
            print(e)
    return wrapper
```

```python
def add(*args, **kwargs):
    sum = 0
    for i in args:
        sum+=i
    for key, value in kwargs.items():
        sum+=value
    print(sum)
```

```python
add = decorator_exep_v2(add)
```

```python
add(1,b=2)
```

# Static

```python
class User:
    # static variable
    course_name = "UDSB2"

    def __init__(self, batch_no):
        self.batch_number = batch_no
```

```python
obj_user_1=User(1);

# static variable can be accessed using class name as well as object name 
# but if we change the value of static variable using object name then it will create a new instance variable with the same name 
# and it will not change the value of static variable for other objects and class name. 
# but if we change the value of static variable using class name then it will change the value of static variable for all objects and class name.
print(obj_user_1.course_name)
print(User.course_name)
obj_user_1.course_name = "hello"
print(obj_user_1.course_name)
print(User.course_name)
User.course_name = "UDSB3"
print(obj_user_1.course_name)
print(User.course_name)
```

# static methods : Used for logical group
- no object required
- cleaner name calling

```python
class MathUtils:

    @staticmethod
    def add(a,b):
        return a + b
    
    @staticmethod
    def subtract_a_from_b(a,b):
        return a - b

    @staticmethod
    def subtract_b_from_a(a,b):
        return b - a
```

```python
MathUtils.add(1,2)
```

```python
MathUtils.subtract_b_from_a(100, 2)
```

```python
class MathUtils_New:
     
    name = "UDSB2"

    @staticmethod
    def add(a,b):
        return a + b
    
    @staticmethod
    def subtract_a_from_b(a,b):
        return a - b

    @staticmethod
    def subtract_b_from_a(a,b):
        return b - a
    
    def class_method(self):
        # instance method can access static method and static variable using class name or self keyword
        print(self.add(1,2))
        print(MathUtils_New.add(1,2))
```

```python
MathUtils_New.add(1,2)
```
