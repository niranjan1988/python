# File Operations

```python
f=open("sample.txt") # Make sure 'sample.txt' exists in the current directory or use full path
content=f.read()
print(content)
f.close() # Always close the file after operations to release resources

```

```python
# to auto manage file closing
with open("sample.txt") as f:
    content=f.read()
    print(content)
# file is automatically closed after the with block
```

```python
f=open("sample.txt")
while True:
    line=f.readline() # read one line at a time
    if not line: # EOF
        break
    print(line, end="")
f.close()
```

```python
f=open("sample.txt")

lines=f.readlines() # read all lines into a list
print(lines) # prints list of lines

for line in lines:
    print(line, end="")
f.close()
```
