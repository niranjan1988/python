# File Operations

```python
f=open("sample.txt") # Make sure 'sample.txt' exists in the current directory or use full path
content=f.read()
print(content)
f.close() # Always close the file after operations to release resources

```

```python
# to auto manage file closing
with open("sample.txt") as f:
    content=f.read()
    print(content)
# file is automatically closed after the with block
```

```python
f=open("sample.txt")
while True:
    line=f.readline() # read one line at a time
    if not line: # EOF
        break
    print(line, end="")
f.close()
```

```python
f=open("sample.txt")

lines=f.readlines() # read all lines into a list
print(lines) # prints list of lines

for line in lines:
    print(line, end="")
f.close()
```

# Read, write and append operations



```python
# Wite starts from 0 index and overwrites existing content. 
# But when we move pointer to a specific index and write, it overwrites from that index onward.
# seek doesnt work with 'a' mode as it always appends at the end.

with open("sample1.txt", "w+") as f:
    initial_index=f.tell() 
    print(initial_index)

    f.write("This is a sample line")
    f.write("\nThis is another line") # pointer is at end of file after first write

    final_index=f.tell() 
    print(final_index)

    mid_index=final_index//2
    print("Middle index ", mid_index)

    f.seek(mid_index) 
    content_from_mid = f.read() 
    print("middle content: ", content_from_mid)

    f.seek(mid_index) 
    f.write("*** ") 
    print("index after appending content", f.tell())
    f.write(content_from_mid)
    
    f.seek(0) # move file pointer to start
    final_content = f.read() # read entire file
    print(final_content)
    

    
```
