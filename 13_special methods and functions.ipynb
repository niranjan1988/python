# Special Methods (OOPS)

Dunder methods are pre-exisitng methods on class to which we assign a behaviour. They name convention looks as __ name __

```python
class Car:
    def __init__(self,manuf, car_name):
        self.manuf = manuf
        self.car_name = car_name

    def __str__(self):
        return f"Manufacturer Name : {self.manuf} and we produce car."
    
    def __repr__(self): # official representation of an object
        return f"Class Car, Manufacturer Name : {self.manuf} and we produce car. This object contains the car model : {self.car_name}"
```

```python
car_1 = Car("toyota", "glanza")

print(car_1) # This will call the __str__ method of the Car class, which provides a user-friendly string representation of the object, suitable for display to end-users.

car_1 # This will call the __repr__ method of the Car class, which provides a detailed and unambiguous string representation of the object, useful for debugging and development purposes.
```

```python
print(str(car_1)) # String represtation of the object
print(f"{car_1!s}") # Both of these will call the __str__ method of the Car class
```

```python
print(repr(car_1)) # Official representation of the object, used for debugging and development purposes.
print(f"{car_1!r}") # __repr__
```

### Next special method - Add

```python
class Wallet:
    def __init__(self, cash):
        self.cash = cash
    
    def __repr__(self):
        return f"Class Wallet, Cash : {self.cash}"
    
    def __add__(self, other): # other is the second operand in the addition operation, which is the second Wallet object being added to the first one.
        return Wallet(self.cash + other.cash)
```

```python
w1 = Wallet(50)
w2 = Wallet(30)
combined_sum = w1 + w2 
print(combined_sum.cash)

print(type(combined_sum))
print(isinstance(combined_sum, Wallet)) # isinstance() checks if the object is an instance of the specified class or a subclass thereof
```

```python
print(isinstance(1, (float,int))) # isinstance() can also check if an object is an instance of any of the types specified in a tuple.
```

### Len and getItem

```python
class ATM:
    def __init__(self):
        self.transacrion_history = []
    
    def __repr__(self):
        return f"Class Wallet, Cash : {self.cash}"
    
    def do_transaction(self, value):
        self.transacrion_history.append(value)
        print(f"Transcation done : {value}")

    def __len__(self):
        return len(self.transacrion_history)
    
    def __getitem__(self, index):
        return self.transacrion_history[index]
    
```

```python
atm_2 = ATM()

atm_2.do_transaction(100)
atm_2.do_transaction(10)
atm_2.do_transaction(-100)
atm_2.do_transaction(84)
```

```python
print(atm_2[3])
len(atm_2)
```

### map

Applied a function to every item in a list

```python
def square_of_number(x):
    return x*x
```

```python
list_a = [1,2,3,4,5]
```

```python
list(map(square_of_number, list_a))
```

### filter

```python
# filter(function, iterable)
# function that takes one item and return a boolean
list_a = [1,2,3,4,5]

def is_even(num):
    # 2 --> 2%2==0 -> True
    # 3 -> 3%2==0 -> False
    return num % 2 == 0
```

```python
list(filter(is_even, list_a))
```

```python
students = {"sharmila" : "batch_1",
            "abhishek" : "batch_2",
            "tanmay": "batch_1",
            "nitin": "batch_2",
            "raj" : "batch_2"}

def check_batch_1(student):
    return student[1] == "batch_1"
```

```python
list(filter(check_batch_1, students.items()))
```

```python
# Example-2
names = ["sharmila","abhishek","tanmay","nitin","raj"]
batch = ['batch_1', "batch_2", "batch_1", "batch_2", "batch_2"]
```

```python
list(zip(names, batch))
```

```python
list(filter(check_batch_1, list(zip(names, batch))))
```
