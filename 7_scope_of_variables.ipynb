# Scope of variables

```python
a=0

def add():
    global a # Use the global variable 'a'
    a +=2
    print(a)

add()
```

```python
a=[1,2,3]
def modify():
    a.append(4) # No need for 'global' since we're modifying a mutable object
    print(a)

modify()
print(a)
```

```python
def add(a,b,c=10,d=20,*args,**kwargs):
    print(a)
    print(b)
    print(c)
    print(d)
    print(args) # Extra positional arguments
    print(kwargs) # Extra keyword arguments

add(1,2,30,40,50,60,x=100,y=200)
```

```python
def add(a=1,b=2,c=10,d=20,*args,**kwargs):
    print(a)
    print(b)
    print(c)
    print(d)
    print(args)
    print(kwargs)

add(c=30) # Only 'c' is provided, others take default values
add(input("Enter a value for 'a': "))
```
